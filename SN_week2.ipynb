```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the graph from the edge list file
G = nx.read_edgelist('D:\\SN\\facebook_combined.txt')
print(nx.info(G))  # Basic information about the graph
print("Number of edges:", nx.number_of_edges(G))
print("Number of nodes:", nx.number_of_nodes(G))
print("Is directed:", nx.is_directed(G))

# Optional: Calculate additional properties
print("Is connected:", nx.is_connected(G))
if nx.is_connected(G):
    print("Diameter of graph:", nx.diameter(G))
print("Average clustering coefficient:", nx.average_clustering(G))

# Visualize a subgraph for clarity
# Extract the largest connected component for a focused visualization
largest_cc = max(nx.connected_components(G), key=len)
subgraph = G.subgraph(largest_cc)

plt.figure(figsize=(10, 10))
nx.draw(subgraph, node_size=10, edge_color="gray", node_color="blue", with_labels=False)
plt.title("Largest Connected Component of Facebook Graph")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_deg_dist(G):
    # Convert DegreeView to a dictionary and get the degree values
    all_degrees = dict(nx.degree(G)).values()
    
    # Get unique degrees
    unique_degrees = list(set(all_degrees))
    
    # Count how many times each unique degree occurs
    count_of_degrees = []
    for degree in unique_degrees:
        count_of_degrees.append(list(all_degrees).count(degree))
    
    # Plot the degree distribution
    plt.plot(unique_degrees, count_of_degrees, 'bo-')
    plt.xlabel('Degree')
    plt.ylabel('Frequency')
    plt.title('Degree Distribution')
    plt.show()

# Call the function with the graph G
plot_deg_dist(G)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx

# Adds nodes to the graph
def add_nodes(n):
    G = nx.Graph()
    G.add_nodes_from(range(n))
    return G

# Adds edges randomly until the graph becomes connected
def add_till_connectivity(G):
    nodes = list(G.nodes())
    while not nx.is_connected(G):
        u, v = np.random.choice(nodes, 2, replace=False)
        G.add_edge(u, v)
    return G

# Creates an instance of the entire process. It takes as input number of nodes and returns the number of edges required for connectivity
def create_instance(n):
    G = add_nodes(n)
    G = add_till_connectivity(G)
    return G.number_of_edges()

# Averages it over 4 instances
def create_avg_instance(n):
    list1 = []
    for i in range(4):
        list1.append(create_instance(n))
    return np.average(list1)

# Plot the desired graph for different number of nodes
def plot_connectivity():
    x = []
    y = []
    i = 10  # i is the number of nodes
    while i <= 400:
        x.append(i)
        y.append(create_avg_instance(i))
        i += 10
    plt.xlabel('Number of nodes')
    plt.ylabel('Number of edges required for connectivity')
    plt.title('Connectivity plot')
    plt.plot(x, y)
    plt.show()

plot_connectivity()
```
